```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
#df se usa para imprimir la base de datos
```

```python
#Verificamos información del DataFrame
df.info()
```

Filtro 1: mostrar los datos del vendedor Leticia Ramirez 

```python
#filtro por objeto
filtro1=df[(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro1
```

crear su .csv 

```python
filtro1.to_csv("Enrtegable1.csv")
```

Filtro 2: mostrar filas no consecutivas 

```python
#filtro por filas no consecutivas 
filtro2= df.iloc[ [2,5], : ]       
filtro2
```

se crea su .csv

```python
filtro2.to_csv("Enrtegable2.csv")
```

Filtro 3: imprimir columnas consecutivas

Filtro 4: imprimir filas y columnas específicas

Filtro 7: MARGEN_TOTAL_CALCULADO menores a 260

Filtro 8: se muestra PRECIO_UNITARIO < 300 y el NOMBRE_VENDEDOR == "LETICIA"

```python
#filtro Y
filtro8 = df[(df["PRECIO_UNITARIO"] < 300) & (df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro8
```

se crea su .csv

```python
filtro8.to_csv("Entregable8.csv")
```

Filtro 9: se muestra "SUBTOTAL_PARTIDA" > 500 o "FECHA_DOC" que sea igual a "2022-12-23"

```python
filtro9 = df[(df['SUBTOTAL_PARTIDA'] > 500) | (df['FECHA_DOC'] == "2022-12-23")]
filtro9
```

se crea su .csv

```python
filtro9.to_csv("Entregale9.csv")
```

Filtro 10: imprimir usando el operador NOT
